## 0) Setup

```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

## 1–3) Scatter plot with color by species
Create a scatter plot with:
- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
# TODO: Create the scatter plot here

plt.figure(figsize=(8,6))
sns.scatterplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    palette="deep"
)
plt.title("Penguin Bill Dimensions by Species")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()

```

**Your interpretation (2–3 sentences):**

We can see distinct clusters by species.
Gentoo penguins tend to have longer but shallower bills, while Adelie have shorter, deeper bills, and Chinstrap fall in between.

## 4) Histogram with two different bin sizes
Plot a histogram of `bill_length_mm` **twice**:
- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.

```python
plt.figure(figsize=(12,5))

# 10 bins

plt.hist(penguins["bill_length_mm"].dropna(), bins=10, color="skyblue", edgecolor="black")
plt.title("Bill Length (10 bins)")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Frequency")

plt.show()
```

```python
# 30 bins
plt.hist(penguins["bill_length_mm"].dropna(), bins=30, color="salmon", edgecolor="black")
plt.title("Bill Length (30 bins)")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

## 5) Boxplot of body mass by species
Create a **boxplot** showing `body_mass_g` grouped by `species`.
Then write 1–2 sentences summarizing which species are heavier/lighter on average.

```python
# TODO: Boxplot of body mplt.figure(figsize=(8,6))
sns.boxplot(
    data=penguins,
    x="species",
    y="body_mass_g",
    palette="pastel"
)
plt.title("Penguin Body Mass by Species")
plt.xlabel("Species")
plt.ylabel("Body Mass (g)")
plt.grid(True, axis="y", linestyle="--", alpha=0.5)
plt.show()

```

**Your summary (1–2 sentences):**
